## Univariate Exploration

> In this section, investigate distributions of individual variables. If
you see unusual points or outliers, take a deeper look to clean things up
and prepare yourself to look at relationships between variables.


the bar and pie charts will be used frequently in this section. so let's define a function for each one.

```python
def barchart(serias,df,size,rotation):
    base_color = sb.color_palette()[0]

    # gets each type with its count
    na_counts = df[serias].value_counts()

    #set appropriate figure size
    plt.subplots(figsize=size)


    sb.countplot( data=df , x=serias, color=base_color , order = na_counts.index);


 

    # printing the count on the top of each bar
    for i in range (na_counts.shape[0]):
        count = na_counts[i]

    
        plt.text(i, count+10, count, ha = 'center')
    
        plt.xticks(rotation=rotation) ;
```

```python
def piechart(serias):
    sorted_counts = df[serias].value_counts()

    plt.subplots(figsize=(7,7))
    plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False,autopct='%1.1f%%', shadow=True);

```

## question 6 :what is the distribution of duration(in second) for all individual rides ? 

```python
#find the minimum value 
minimum = df["duration_min"].min()

#find the maximum value
maximum = df["duration_min"].max()

# calculate the step value to make 40 bins
step = (maximum-minimum)/40

bins =np.arange(minimum , maximum + step, step);

plt.hist(data=df,x="duration_min",bins=bins);

plt.xlabel('durations (minuts)');
```

>this issue is discussed below,however ,we will perform a log transformations to solve this problemlow,howefwr ,we will perform a log transformations

```python
minimum = math.log10(minimum) 

maximum = math.log10(maximum) 

step = (maximum-minimum)/40

bins = 10 ** np.arange(minimum, maximum+step, step)

ticks = [ 1,2, 3,10,20, 30, 100, 300, 600,1410]

labels = ['{}'.format(val) for val in ticks]

plt.hist(data = df, x = 'duration_min', bins = bins);

plt.xscale('log');

plt.xticks(ticks, labels);

plt.xlabel('durations (minuts)');

```

<span style="color:blue">Observations: the data looked right-skewed a little pit with one peak between 3 and 25 minutes. duration is approximately bounded between 3 to 300 minutes . </span>

```python
def Adapted_Bar_Chart(x,y,xlabel,ylabel,title):


    base_color = sb.color_palette()[0]

    sb.barplot(data=df, x=x, y=y, color=base_color)

    plt.ylabel(ylabel);
    plt.xlabel(xlabel);
    plt.title(title);
```

```python
def boxplot (x,y,xlabel,ylabel):
    sb.boxplot(data=df, y=y, x=x, color=base_color);
    plt.ylabel(ylabel);
    plt.xlabel(xlabel);
```
